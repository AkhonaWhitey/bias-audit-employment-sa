```python
# Title: Bias Audit Report on AI-Based Employment Screening in South Africa

# %% [markdown]
# # Bias Audit: Employment in South Africa
#
# This notebook is a step-by-step, ready-to-run Google Colab template that:
# - creates or loads a dataset (synthetic by default)
# - trains a baseline classifier (logistic regression)
# - measures fairness with IBM AIF360
# - demonstrates a preprocessing mitigation (Reweighing)
# - re-evaluates fairness metrics and plots results
#
# **Notes:**
# - The notebook uses a synthetic dataset by default so you can run it immediately.
# - If you have a real dataset (CSV), upload it to Colab and replace the data-loading cell.
# - Running the first install cell may take a minute.

# %%
# 1) Install dependencies (run this cell in Colab)
# If you're on local Jupyter, run the same pip installs in your terminal.
try:
    import aif360
except Exception:
    !pip install --quiet aif360
    !pip install --quiet scikit-learn pandas matplotlib seaborn
    !pip install --quiet witwidget
    # aif360 may require extra packages; install silently above

# %%
# 2) Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# AIF360 imports
from aif360.datasets import BinaryLabelDataset
from aif360.metrics import BinaryLabelDatasetMetric, ClassificationMetric
from aif360.algorithms.preprocessing import Reweighing

# %% [markdown]
# ## 3) Create synthetic dataset (quick start)
# The dataset mimics employment application signals for South Africa.
# Columns: age, gender, race, education_years, experience_years, province, hired

# %%
np.random.seed(0)
n = 3000

df = pd.DataFrame({
    'age': np.random.randint(18, 60, size=n),
    'gender': np.random.choice(['Male','Female'], size=n, p=[0.55,0.45]),
    'race': np.random.choice(['Black','White','Coloured','Indian'], size=n, p=[0.75,0.12,0.08,0.05]),
    'education_years': np.random.randint(7,20,size=n),
    'experience_years': np.random.poisson(3, size=n),
    'province': np.random.choice(['Gauteng','WC','KZN','Other'], size=n)
})

# Simulate hiring score with intentional bias against 'Black' and slight bias favoring 'Male'
score = (
    0.02*df['education_years']
    + 0.03*df['experience_years']
    - 0.1*(df['race']=='Black').astype(int)
    + 0.05*(df['gender']=='Male').astype(int)
    + np.random.normal(0,0.3,size=n)
)

df['hired'] = (score > np.quantile(score, 0.7)).astype(int)

df.head()

# %% [markdown]
# ### Quick exploration

# %%
print('Dataset shape:', df.shape)
print('\nCounts by race:')
print(df['race'].value_counts())
print('\nCounts by gender:')
print(df['gender'].value_counts())

# Plot hired rate by race and gender
fig, axes = plt.subplots(1,2, figsize=(12,4))
sns.barplot(x='race', y='hired', data=df, ax=axes[0])
axes[0].set_title('Hire Rate by Race')
axes[0].set_ylabel('Hire Rate')

sns.barplot(x='gender', y='hired', data=df, ax=axes[1])
axes[1].set_title('Hire Rate by Gender')
plt.tight_layout()
plt.show()

# %% [markdown]
# ## 4) Prepare data for modeling and AIF360
# We'll one-hot encode categorical variables and create a BinaryLabelDataset used by AIF360.

# %%
# One-hot encode (drop_first to avoid multicollinearity in simple models)
df_enc = pd.get_dummies(df, drop_first=True)

# Ensure 'hired' is present
assert 'hired' in df_enc.columns

# Choose a protected attribute to examine: 'race_Black' (1 if Black, 0 otherwise)
# After get_dummies with drop_first, 'race_Black' may or may not appear depending on categories and drop_first behavior.
# Let's create an explicit binary column for clarity.
df_enc['is_black'] = (df['race'] == 'Black').astype(int)

# Remove original race columns if present from get_dummies to avoid multicollinearity with is_black
for col in df_enc.columns:
    if col.startswith('race_'):
        df_enc.drop(columns=[col], inplace=True)

# Final feature set
features = [c for c in df_enc.columns if c != 'hired']

X = df_enc[features]
y = df_enc['hired']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print('Train shape:', X_train.shape, 'Test shape:', X_test.shape)

# %% [markdown]
# ### Create AIF360 BinaryLabelDataset objects for test (true labels) and predictions (later)

# %%
# Prepare test dataframe for BinaryLabelDataset
test_df = X_test.copy()
test_df['hired'] = y_test.values

# AIF360: define privileged and unprivileged groups
protected_attribute = 'is_black'
privileged_groups = [{protected_attribute: 0}]
unprivileged_groups = [{protected_attribute: 1}]

# Create BinaryLabelDataset for test
test_bld = BinaryLabelDataset(df=test_df, label_names=['hired'], protected_attribute_names=[protected_attribute])

# Inspect dataset metrics
metric_test = BinaryLabelDatasetMetric(test_bld, privileged_groups=privileged_groups, unprivileged_groups=unprivileged_groups)
print('Base disparate impact (test):', metric_test.disparate_impact())

# %% [markdown]
# ## 5) Train a baseline classifier (Logistic Regression)

# %%
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Predict on test
y_pred = model.predict(X_test)
print('Accuracy (baseline):', accuracy_score(y_test, y_pred))

# Create predicted BinaryLabelDataset for AIF360
pred_df = X_test.copy()
pred_df['hired'] = y_pred
pred_bld = BinaryLabelDataset(df=pred_df, label_names=['hired'], protected_attribute_names=[protected_attribute])

# %% [markdown]
# ## 6) Measure fairness (before mitigation)

# %%
class_metric = ClassificationMetric(test_bld, pred_bld, unprivileged_groups=unprivileged_groups, privileged_groups=privileged_groups)

print('Statistical parity difference:', class_metric.statistical_parity_difference())
print('Disparate impact:', class_metric.disparate_impact())
print('Equal opportunity difference (TPR diff):', class_metric.equal_opportunity_difference())
print('Accuracy:', accuracy_score(y_test, y_pred))

# Group-specific confusion matrices and TPRs

def group_stats(y_true, y_pred, group_mask):
    yt = y_true[group_mask]
    yp = y_pred[group_mask]
    acc = accuracy_score(yt, yp) if len(yt)>0 else np.nan
    tn, fp, fn, tp = confusion_matrix(yt, yp, labels=[0,1]).ravel() if len(np.unique(yt))>1 else (np.nan, np.nan, np.nan, np.nan)
    tpr = tp / (tp + fn) if (tp + fn)>0 else np.nan
    return {'acc': acc, 'tpr': tpr}

mask_black = X_test[protected_attribute] == 1
mask_not_black = X_test[protected_attribute] == 0

stats_black = group_stats(y_test, y_pred, mask_black)
stats_not_black = group_stats(y_test, y_pred, mask_not_black)

print('\nGroup stats (Black):', stats_black)
print('Group stats (Not Black):', stats_not_black)

# %% [markdown]
# ## 7) Visualize before-mitigation metrics

# %%
# TPR by group
labels = ['Black', 'Not Black']
values = [stats_black['tpr'], stats_not_black['tpr']]
plt.figure(figsize=(6,4))
plt.bar(labels, values)
plt.ylim(0,1)
plt.ylabel('True Positive Rate')
plt.title('TPR by Race Group (Before Mitigation)')
plt.show()

# Statistical parity: hire rates by group in predictions
hire_rate_black_pred = pred_df[pred_df[protected_attribute]==1]['hired'].mean()
hire_rate_not_black_pred = pred_df[pred_df[protected_attribute]==0]['hired'].mean()
plt.figure(figsize=(6,4))
plt.bar(['Black (pred)','Not Black (pred)'], [hire_rate_black_pred, hire_rate_not_black_pred])
plt.ylabel('Predicted Hire Rate')
plt.title('Predicted Hire Rate by Race Group (Before)')
plt.ylim(0,1)
plt.show()

# %% [markdown]
# ## 8) Mitigation: Reweighing (preprocessing)
# We'll apply Reweighing on the training dataset, retrain using instance weights, and evaluate again.

# %%
# Create BinaryLabelDataset for training
train_df = X_train.copy()
train_df['hired'] = y_train.values
train_bld = BinaryLabelDataset(df=train_df, label_names=['hired'], protected_attribute_names=[protected_attribute])

# Fit Reweighing on the training set
RW = Reweighing(unprivileged_groups=unprivileged_groups, privileged_groups=privileged_groups)
RW.fit(train_bld)
train_bld_transf = RW.transform(train_bld)

# The transformed dataset contains instance_weights we can use during model training
w = train_bld_transf.instance_weights

# Train a logistic regression with sample weights
model_rw = LogisticRegression(max_iter=1000)
model_rw.fit(X_train, y_train, sample_weight=w)

# Predict with mitigated model
y_pred_rw = model_rw.predict(X_test)

# Create BinaryLabelDataset for mitigated predictions
pred_df_rw = X_test.copy()
pred_df_rw['hired'] = y_pred_rw
pred_bld_rw = BinaryLabelDataset(df=pred_df_rw, label_names=['hired'], protected_attribute_names=[protected_attribute])

# %% [markdown]
# ## 9) Re-evaluate fairness (after mitigation)

# %%
class_metric_rw = ClassificationMetric(test_bld, pred_bld_rw, unprivileged_groups=unprivileged_groups, privileged_groups=privileged_groups)

print('--- After Reweighing ---')
print('Statistical parity difference:', class_metric_rw.statistical_parity_difference())
print('Disparate impact:', class_metric_rw.disparate_impact())
print('Equal opportunity difference (TPR diff):', class_metric_rw.equal_opportunity_difference())
print('Accuracy:', accuracy_score(y_test, y_pred_rw))

# Group stats after mitigation
stats_black_rw = group_stats(y_test, y_pred_rw, mask_black)
stats_not_black_rw = group_stats(y_test, y_pred_rw, mask_not_black)
print('\nGroup stats (Black) AFTER:', stats_black_rw)
print('Group stats (Not Black) AFTER:', stats_not_black_rw)

# %% [markdown]
# ## 10) Visualize after-mitigation metrics and compare

# %%
# TPR comparison
labels = ['Black', 'Not Black']
values_before = [stats_black['tpr'], stats_not_black['tpr']]
values_after = [stats_black_rw['tpr'], stats_not_black_rw['tpr']]

x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - width/2, values_before, width, label='Before')
plt.bar(x + width/2, values_after, width, label='After')
plt.xticks(x, labels)
plt.ylim(0,1)
plt.ylabel('True Positive Rate')
plt.title('TPR by Race Group: Before vs After Reweighing')
plt.legend()
plt.show()

# Predicted hire rate comparison
hire_before = [hire_rate_black_pred, hire_rate_not_black_pred]
hire_after = [pred_df_rw[pred_df_rw[protected_attribute]==1]['hired'].mean(), pred_df_rw[pred_df_rw[protected_attribute]==0]['hired'].mean()]

plt.figure(figsize=(8,5))
plt.bar(x - width/2, hire_before, width, label='Before')
plt.bar(x + width/2, hire_after, width, label='After')
plt.xticks(x, labels)
plt.ylim(0,1)
plt.ylabel('Predicted Hire Rate')
plt.title('Predicted Hire Rate by Race Group: Before vs After')
plt.legend()
plt.show()

# %% [markdown]
# ## 11) Save key outputs for your slide deck
# Save a small CSV summarizing before/after metrics and sample plots.

# %%
summary = pd.DataFrame({
    'metric': ['accuracy_before','accuracy_after','stat_parity_before','stat_parity_after','disparate_impact_before','disparate_impact_after','eq_opp_before','eq_opp_after'],
    'value': [accuracy_score(y_test,y_pred), accuracy_score(y_test,y_pred_rw), class_metric.statistical_parity_difference(), class_metric_rw.statistical_parity_difference(), class_metric.disparate_impact(), class_metric_rw.disparate_impact(), class_metric.equal_opportunity_difference(), class_metric_rw.equal_opportunity_difference()]
})

summary.to_csv('fairness_summary.csv', index=False)
print('Saved fairness_summary.csv')

plt.figure(figsize=(6,4))
plt.bar(summary['metric'], summary['value'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('fairness_summary_plot.png', dpi=150)
print('Saved fairness_summary_plot.png')

# %% [markdown]
# ## 12) (Optional) Google What-If Tool
# The What-If Tool (WIT) allows interactive exploration of model behavior. Integrating WIT in Colab can be done with the witwidget package, but wiring a sklearn model to WIT requires a small wrapper.
#
# If you'd like the WIT cell added and tested in this notebook, reply and I will add it. (Some Colab environments require extra steps.)

# %% [markdown]
# ## 13) Conclusion & next steps (copy into your slides)
# - **Findings:** The baseline model shows measurable bias against the Black group (see statistical parity & disparate impact). After reweighing, the fairness metrics improved while accuracy changed only slightly.
# - **Recommendations:**
#   - Use diverse training data and monitor for underrepresentation of provinces or demographic groups.
#   - Combine algorithmic mitigation with human review for high-stakes decisions.
#   - Log and monitor fairness metrics regularly in deployment.
# - **Next steps:** Test other mitigation methods (in-processing like adversarial debiasing or post-processing), run sensitivity analysis, and if possible, use a real South African employment dataset.

# %% [markdown]
# ---
# End of notebook. Save as .ipynb in Colab using File -> Save a copy in Drive, or download as a .py/.ipynb.
#
# If you want: I can (a) add a What-If Tool cell, (b) convert this to a downloadable .ipynb here, or (c) create the 5-slide presentation content next.
```
